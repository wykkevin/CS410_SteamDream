```python
from google.colab import drive
import pandas as pd
import os

drive.mount('/content/drive',force_remount=True)
folder_path = './drive/MyDrive/Colab Notebooks/CS410/data/processed_data/'
output_file = './drive/MyDrive/Colab Notebooks/CS410/data/merged.csv'
```

```python
import pandas as pd
import os

def merge_csv(folder_path, output_file):
    merged_df = pd.DataFrame()

    for file in os.listdir(folder_path):
        if file.endswith('.csv'):
            file_path = os.path.join(folder_path, file)
            df = pd.read_csv(file_path)
            merged_df = pd.concat([merged_df, df], ignore_index=True)

    merged_df.to_csv(output_file, index=False)


merge_csv(folder_path, output_file)
```
